```python
lab = list(df_crime['Group_Name'].unique())
chart = sns.countplot(df_crime["Group_Name"])
chart.set_xticklabels(labels=lab, rotation=90)
```

```python
plt.figure(figsize=(15,10))
lab = list(df_crime["Area_Name"].unique())
chart = sns.countplot(df_crime["Area_Name"])
chart.set_xticklabels(labels=lab, rotation=90)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
columns=["Area_Name","Group_Name"]
for val in columns:
    df_crime[val]=le.fit_transform(df_crime[val].astype(str))
df_crime
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x="Area_Name",y="Group_Name",data=df_crime)
```
